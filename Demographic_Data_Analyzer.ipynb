```python
4.#percentage of advanced education above 50k
advanced_edu = df.groupby(['education', 'over50k']).sum().iloc[18:22,:] / df.groupby(['education']).sum() * 100
advanced_edu.iloc[::-2]
```

```python
5. #percentage of non-advanced education above 50k
non_advanced_edu = df.groupby(['education', 'over50k']).sum() / df.groupby(['education']).sum() * 100 
non_advanced_edu = non_advanced_edu.sort_values(by='over50k', ascending=False).iloc[:15,:]
non_advanced_edu
```

```python
test=non_advanced_edu.unstack() #unstack the multi index to get the education index only
test=test.drop(test[test.index.isin([' Masters',' Doctorate',' Bachelors'])].index) #drop the selected indices
test.stack() #restack the data 
```

```python
# Minimum number of hours a person work per weeek
df['hoursperweek'].min()
```

```python
# Percentage of people that work minimum hour above 50K
min_percent = df.groupby(['hoursperweek', 'over50k']).sum() / df.groupby(['hoursperweek']).sum() * 100
min_percent.iloc[1:2,:]
```

```python
# Country with the highest number of people that earn above 50K
country_50 = df[df.over50k == ' >50K']  # display country over 50K
country_50.nativecountry.value_counts()[[0]]
```

```python
# most popular occupation in India above 50K
country_grp=df.groupby('nativecountry')   #groupby country 
india_grp=country_grp.get_group(' India') #get all India 
india_50k=india_grp[india_grp['over50k']== ' >50K']  #display country over 50K
india_50k.occupation.value_counts()[[0]]
```
